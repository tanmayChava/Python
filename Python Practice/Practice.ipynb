Q1 Swap two variables without using a third variable:

```python
a = 10
b = 20
a,b = b,a
print(f"After Swapping a={a},b={b}")
```

```python
a = 10
b = 20
temp = a
a = b
b = temp
print(f"After Swapping a={a},b={b}")
```

Q6 Create a list of odd numbers between 11-30:

```python
odd_number = []
for i in range (11,20):
    if i % 2!=0 :
        odd_number.append(i)
print(odd_number)
```
