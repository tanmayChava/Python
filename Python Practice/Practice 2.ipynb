```python
Q18 Program to show multiple inheritance:
```

```python
class Animal:
    def speak(self):
        print("Animal make Sound")

class Dog(Animal):
    def bark(self):
        print("Dog Bark")

class Cat(Animal):
    def meow(self):
        print("Cat Meow")

class Pet(Dog,Cat):
    pass

pet = Pet()
pet.speak()
pet.bark()
pet.meow()
```
