```python
Q35 Function for Reversed String
```

```python
def reversed_string(s):
    return s[::-1]
print(reversed_string("Hello"))
```

```python
# In-Built Function
def reversed_string_join(s):
    return "".join(reversed(s))
    
string = "World"
reversed_string_1 = reversed_string_join(string)
print(reversed_string_1)
```

```python
Q35 Function to count vowels in a string
```

```python
def find_vowels(s):
    vowels = 'AEIOUaeiou'
    return sum(1 for char in s if char in vowels )

print(find_vowels("HelloWorld"))
```

```python
def find_vowels(s):
    vowels = 'aeiou'
    count = 0

    for char in s.lower() :
        if char in vowels:
            count +=1
    return count

print(find_vowels("Hello World"))
```

```python
def vowel_counts(s):
    vowels = "aeiou"
    s = s.lower()
    result = {}

    for v in vowels:
        count = s.count(v)
        if count > 0:
            result[v] = count

    return result

print(vowel_counts("Hello World")) 
```

```python
Q38 Count occurrences of unique characters and sort (descending)
```

```python
from collections import Counter

def count_unique_chars(s):
    counter = Counter(s)
    sorted_count = sorted(counter.items(), key=lambda x: x[1], reverse=True)
    return sorted_count

print(count_unique_chars("aabbccddeeff"))
```
