```python
Question: How can you flatten a nested list (list of lists) into a single list?
```

```python
nested_list = [[1, 2], [3, 4], [5, 6]]
flattened_list = []
for sub_list in nested_list:
    for item in sub_list:
        flattened_list.append(item)
print(flattened_list)
        
```

```python
Question: How can you find the most frequent element in a list?
```

```python
from collections import Counter
my_list = [1, 2, 2, 3, 3, 3, 4]
count = Counter(my_list)
most_common = count.most_common(1)[0][0]
print(most_common)
```

```python
Question: How do you remove duplicates from a list while preserving the order of elements?
```

```python
def remove_duplicate(my_list):
    seen = set()
    result = []

    for item in my_list:
        if item not in seen:
            seen.add(item)
            result.append(item)
    return result

my_list = [1, 2, 2, 3, 4, 4, 5]
list2 = remove_duplicate(my_list)
print(list2)        
```

```python
Question: How would you rotate a list by n positions to the right or left?
```

```python
my_list = [1, 2, 3, 4, 5]
n = 2
rotated_list = my_list[-n:] + my_list[:-n]
print(rotated_list)
```

```python
rotated_list = my_list[n:] + my_list[:n]
print(rotated_list)
```

```python
Question: How can you find the intersection of multiple lists?
```

```python
list1 = [1, 2, 3, 4]
list2 = [3, 4, 5, 6]
list3 = [4, 5, 6, 7]

intersection = list(set(list1) & set(list2) & set(list3))
print(intersection)
```
