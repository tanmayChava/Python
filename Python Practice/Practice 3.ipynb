```python
fruits = ["Apple", "Banana", "Cherry", "Date", "Elderberry"]
second_letter = []

for fruit in fruits:
    if len(fruit) > 2:
        second_letter.append(fruit[1])
    else:
        second_letter.append(None)
print(second_letter)
```

```python
fruits = ["Apple", "Banana", "Cherry", "Date", "Elderberry"]
second_letter = [fruit[1] for fruit in fruits]
print(second_letter)
```

```python
Q32. Python program to check if a year is a leap year
```

```python
year = int(input("Enter the Year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap Year")
else:
    print(f"{year} is a not leap Year")
```
